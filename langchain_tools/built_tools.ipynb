custom tool

```python
from langchain_core.tools import tool
@tool
def multiply(a: int, b: int) -> int:
    """Multiplies two numbers together."""
    return a * b
  
```

```python
multiply.invoke({"a": 5, "b": 6})
```

# custom funcation with pydantic and struture tool

```python
from pydantic import BaseModel, Field
from langchain_core.tools import StructuredTool
```

```python
class multiple_input(BaseModel):
    a: int = Field(required=True, description="the first number")
    b: int = Field(required=True, description="the second number")
```

```python
def multiply(a: int, b: int) -> int:
    """Multiplies two numbers together."""
    return a * b
```

```python
structured_tool=StructuredTool.from_function(
    func=multiply,
    name="multiply",
    description="Multiplies two numbers together.",
    args_schema=multiple_input
)
```

```python
structured_tool.invoke({"a": 5, "b": 10})
```

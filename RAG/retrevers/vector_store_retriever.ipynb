```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings,HuggingFaceEndpoint
from dotenv import load_dotenv
```

```python
embeddings=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2" 
)
```

```python
from langchain_core.documents import Document
```

```python
docs = [
    Document(page_content="""
Machine Learning (ML) is a subset of Artificial Intelligence that enables systems to learn patterns from data 
and make predictions without being explicitly programmed. Common algorithms include linear regression, decision trees, 
and neural networks.
"""),

    Document(page_content="""
Deep Learning is a specialized branch of machine learning that uses neural networks with multiple layers. 
It is widely used in computer vision, natural language processing, and speech recognition tasks.
"""),

    Document(page_content="""
LangChain is a framework designed to build applications powered by Large Language Models (LLMs). 
It allows developers to combine LLMs with external data sources, APIs, and tools to create intelligent systems.
"""),

    Document(page_content="""
Retrieval-Augmented Generation (RAG) is a technique that improves LLM responses by retrieving relevant information 
from external sources before generating an answer. It combines retrieval systems with generative models.
"""),

    Document(page_content="""
Vector databases such as FAISS and Chroma store embeddings of text data, enabling efficient similarity search. 
These databases are essential for building scalable RAG systems and semantic search applications.
""")
]
```

```python
vector_stores=FAISS.from_documents( docs,embeddings)
```

```python
vector_stores.similarity_search(
    query = "What is the role of China in Pakistan and India relations?",
    k=1
)
```

```python
# convert vector into retriever
retriever = vector_stores.as_retriever(search_kwargs={"k": 2})
```

```python
retriever1=vector_stores.as_retriever(
    search_type="mmr",
    search_kwargs={"k":2, "lambda_mult":0}
                                      )
```

```python
query3 = "What is deep learning?"
result=retriever1.invoke(query3)
```

```python
for i , doc in enumerate(result):
    print(f"\n---result {i+1}---\n")
    print(f"content :\n {doc.page_content}")
```

```python
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
from langchain_groq import ChatGroq
from langchain_huggingface import HuggingFaceEmbeddings,HuggingFaceEndpoint
from dotenv import load_dotenv
```

```python
embeddings=HuggingFaceEmbeddings(   
     model_name="sentence-transformers/all-MiniLM-L6-v2"
      )
```

```python
from langchain_core.documents import Document

docs = [
    Document(
        page_content="Machine learning and deep learning enable systems to learn patterns from data and make intelligent predictions in tasks like image recognition and NLP. In contrast, ancient Roman architecture focused on durability, using stone and concrete to build long-lasting structures like aqueducts and amphitheaters.",
        metadata={"topic": "ml_dl"}
    ),

    Document(
        page_content="LangChain and large language models help developers build AI applications by combining reasoning with external data sources and APIs. Meanwhile, ocean currents play a key role in regulating Earth's climate by distributing heat across different regions of the planet.",
        metadata={"topic": "langchain_llm"}
    ),

    Document(
        page_content="Vector databases and embeddings allow efficient similarity search, making them essential for semantic search and RAG systems. In contrast, traditional music theory studies harmony, rhythm, and melody to compose structured musical pieces across genres.",
        metadata={"topic": "vector_db_embeddings"}
    ),

    Document(
        page_content="Retrieval-Augmented Generation improves AI responses by combining document retrieval with text generation for more accurate answers. On the other hand, volcanic eruptions release magma, ash, and gases from beneath Earth's crust, shaping landscapes over time.",
        metadata={"topic": "rag_system"}
    ),

    Document(
        page_content="AI applications like chatbots and recommendation systems use machine learning and user data to provide personalized experiences. Similarly, ancient agricultural practices relied on seasonal cycles and irrigation systems to grow crops efficiently.",
        metadata={"topic": "ai_applications"}
    )
]
```

```python
model=ChatGroq(
    model="llama-3.3-70b-versatile"
)
```

```python
vector_stores=FAISS.from_documents(
    docs,embeddings
)
```

```python
similarity_retrivers=vector_stores.as_retriever(
    search_kwargs={"k":3}
)
```

```python
from langchain_classic.retrievers.document_compressors import LLMChainExtractor
from langchain_classic.retrievers import ContextualCompressionRetriever

compressor = LLMChainExtractor.from_llm(model)
```

```python
compression_retriever=ContextualCompressionRetriever(
    base_retriever=vector_stores.as_retriever(
    search_kwargs={"k":3}
),
base_compressor=compressor

)
```

```python
query = "What is retrieval augmented generation and how does it work?"
result=compression_retriever.invoke(query)
```

```python
for i, doc in enumerate(result):
    print(f"\n---result {i+1}---\n")
    print(f"content :\n{doc.page_content}")
```

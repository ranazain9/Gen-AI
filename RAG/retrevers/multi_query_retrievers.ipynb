```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings,HuggingFaceEndpoint
from langchain_classic.retrievers.multi_query import MultiQueryRetriever
from dotenv import load_dotenv
```

```python
embeddings=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2" 
)
```

```python
from langchain_core.documents import Document

docs = [
    Document(
        page_content="Exercise improves physical and mental health by controlling weight, reducing chronic disease risk, and lowering stress.",
        metadata={"topic": "exercise", "category": "physical_health"}
    ),

    Document(
        page_content="A balanced diet with fruits, vegetables, whole grains, and proteins supports immunity and prevents malnutrition.",
        metadata={"topic": "diet", "category": "nutrition"}
    ),

    Document(
        page_content="Mental health includes emotional and psychological well-being. Therapy, medication, and lifestyle changes help manage anxiety and depression.",
        metadata={"topic": "mental health", "category": "wellbeing"}
    ),

    Document(
        page_content="Sleep is essential for recovery and brain function. Adults need 7–9 hours of sleep for cognitive and emotional stability.",
        metadata={"topic": "sleep", "category": "lifestyle"}
    ),

    Document(
        page_content="Preventive healthcare involves checkups, vaccinations, and screenings to detect diseases early and improve outcomes.",
        metadata={"topic": "prevention", "category": "healthcare"}
    ),

    Document(
        page_content="Hydration supports body temperature, joints, and energy levels. Drinking enough water improves focus and overall function.",
        metadata={"topic": "hydration", "category": "lifestyle"}
    ),

    Document(
        page_content="Chronic diseases like diabetes and hypertension require lifestyle changes, medication, and regular monitoring.",
        metadata={"topic": "chronic disease", "category": "medical"}
    )
]
```

```python
vector_stores=FAISS.from_documents(
    docs,embeddings
)
```

```python
from langchain_groq import ChatGroq
```

```python
model=ChatGroq(
    model="llama-3.3-70b-versatile"
)
```

```python
retriver1=vector_stores.as_retriever(
    search_type="similarity",
    search_kwags={"k":5}

)
```

```python
multi_query=retriver=MultiQueryRetriever.from_llm(
    retriever= vector_stores.as_retriever(
    search_kwags={"k":5}
  ),
  llm=model

)
```

```python
query="how to stay healthy"
result1=retriver1.invoke(query)
result2=multi_query.invoke(query)
```

```python
for i, doc in enumerate(result1):
    print(f"\n---result {i+1}---\n")
    print(f"content\n. {doc.page_content}")
```

```python
for i, doc in enumerate(result2):
    print(f"\n---result {i+1}---\n")
    print(f"content\n. {doc.page_content}")
```

```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings,HuggingFaceEndpoint
from dotenv import load_dotenv
from langchain_core.
```

```python
embeddings=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2" 
)
```

```python
from langchain_core.documents import Document
docs = [
    Document(page_content="Pakistan and India have a complex geopolitical history involving wars and diplomacy."),
    Document(page_content="China has played a strategic role in South Asian geopolitics, especially in relation to Pakistan."),
    Document(page_content="India-China relations include border disputes and economic competition."),
    Document(page_content="The China-Pakistan Economic Corridor is a major project connecting both countries."),
    Document(page_content="Geopolitical tensions in South Asia affect global trade and security.")
]
```

```python
vector_stores=FAISS.from_documents( docs,embeddings)
```

```python
retriever1=vector_stores.as_retriever(
    search_type="mmr",
    search_kwargs={"k":2, "lambda_mult":0}
                                      )
```

```python
query3 = "Indea anf pakistan history geopolitical?"
result=retriever1.invoke(query3)
```

```python
for i , doc in enumerate(result):
    print(f"\n---result {i+1}---\n")
    print(f"content :\n {doc.page_content}")
```

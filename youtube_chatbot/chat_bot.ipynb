```python
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled, NoTranscriptFound

video_id = "J5_-l7WIO_w"

api = YouTubeTranscriptApi()

# Try to fetch transcript - first English, then fall back to any available language
try:
    transcript_list = api.fetch(video_id, languages=['en'])
    transcript_text = " ".join([snippet.text for snippet in transcript_list.snippets])
    print("English transcript fetched successfully!")
    print(transcript_text)
except NoTranscriptFound as e:
    # If English not available, try to get any available transcript
    print(f"English transcript not found. Trying Hindi...")
    try:
        transcript_list = api.fetch(video_id, languages=['hi'])
        transcript_text = " ".join([snippet.text for snippet in transcript_list.snippets])
        print("Hindi transcript fetched successfully!")
        print(transcript_text)
    except Exception as fallback_error:
        print(f"Failed to fetch Hindi transcript: {fallback_error}")
        print("No transcript available for this video.")
except Exception as e:
    print(f"Error fetching transcript: {e}")
```
